```python
import pandas as pd
```

```python
data = pd.read_csv("../docword.nips.txt", skiprows =3, sep=" ", header = None, names=['Doc', 'Word', 'Count'])
```

```python
data
```

```python
freq = pd.pivot_table(data,index=['Doc'],columns=['Word'])
```

```python
freq
```

```python
num_words = data['Word'].unique().shape[0]
num_docs = data['Doc'].unique().shape[0]
```

```python
num_docs
```

```python
x = freq.describe()
```

```python
df = x.loc['count',:]
```

```python
freq = freq.fillna(0)
```

```python
freq.head()
```

```python
max1 = freq.max()
```

```python
max1.max()
```

```python
freq = freq/max1.max()
```

```python
idf = 1/df
```

```python
freq = freq.multiply(idf)
```

```python
freq
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 20,tol=0.0001, init = 'random')
```

```python
classes = km.fit_predict(freq)
```

```python
from matplotlib import pyplot as plt
plt.hist(classes,log=True)
```
